## Data normalization

Before running the SVM, it is always desirable to normalize our training data so each feature has zero sample mean and unit variance.

** Question 1 (1 point) **

Normalize your training and test data using the training **sample mean and std**.

```python
# A function that normalizes data with pre-scecified mean and std. 
def normalize(X,mu=0,std=1,flag_train=True):
    if(flag_train):
        mu = np.mean(X,0)
        std = np.std(X,0)
    
    X -= mu
    X /= std
    
    return X,mu,std


img_norm, mu, std = normalize(images_red.astype(np.float64),mu=0,std=1,flag_train=True)
img_test_norm,_,_ = normalize(images_red_test.astype(np.float64),mu,std,flag_train=False)

```

* Apply a 5-times cross validation to estimate the missclasification rate varying C from .01 to 100 (in log scale) with $\gamma$ equal to 1.

```python
def CVforC(x,y,c):
    a=np.zeros([len(c),5])
    for index, i in enumerate(c):
        model = SVC(C=i ,kernel='rbf',  gamma=1)
        k=ShuffleSplit(5, 0.3)
        for train_index, test_index in k.split(x):
            fit = model.fit(x[train_index], y[train_index])
            scores = model.predict(x[test_index])
            a[index,] = accuracy_score(y[test_index],scores)
    return a    
```

* Apply a 5-times cross validation to estimate the missclasification rate varying C and $\gamma$ around the best values obtained previously in steps 2x.


```python
def CVforCandGamma(x,y,c,g):
    a=np.zeros([len(c),len(g)])
    for index, i in enumerate(g):
        for indice, j in enumerate(c):
            model = SVC(C=j,kernel='rbf',  gamma=i)
            k=ShuffleSplit(5, 0.3)
            foldA =[]
            for train_index, test_index in k.split(x):
                fit = model.fit(x[train_index], y[train_index])
                scores = model.predict(x[test_index])
                foldA.append(accuracy_score(y[test_index],scores))
            a[index,indice] = np.mean(foldA)
    return a       
```

```python
c=[5,10,15]
g=[0.5,1,1.5]
accuracyCandGamma=CVforCandGamma(img_norm, labels_red,c,g)
```

```python
def CVforCpoly(x,y,c):
    a=np.zeros([len(c),5])
    for index, i in enumerate(c):
        model = SVC(C=i ,kernel='poly',  gamma=1)
        k=ShuffleSplit(5, 0.3)
        for train_index, test_index in k.split(x):
            fit = model.fit(x[train_index], y[train_index])
            scores = model.predict(x[test_index])
            a[index,] = accuracy_score(y[test_index],scores)
    return a    
```

* Apply a 5-times cross validation to estimate the missclasification rate varying degree from 1 (linear) to 5 and the best C obtained in the previous step.

```python
from sklearn.model_selection import ShuffleSplit
from sklearn.metrics import accuracy_score
def CVforDegree(x,y,d):
    a=np.zeros([len(d),5])
    for index, i in enumerate(d):
        model = SVC(C=100,kernel='poly',gamma=1,degree=i)
        k=ShuffleSplit(5, 0.3)
        for train_index, test_index in k.split(x):
            fit = model.fit(x[train_index], y[train_index])
            scores = model.predict(x[test_index])
            a[index,] = accuracy_score(y[test_index],scores)
    return a    
```

## Doing some feature engineering

So far, you have essentially run your classifier without no feature engineering at all. While this may work, it is not a good idea in general.

** Question 4 (3 points) **

For each category, plot the mean of each features and try to understand if there are subsets of features (or new ones that could be created) where both categories can be separated away in a more effective way. If so, implement the SVC with the new set of features and evaluate the performance compared to the models you run before.

For instance, if we were comparing dogs & frogs, I would expect that the green levels in frog images would be higher than in dogs images.

**Note** There is no closed solution to this. I just want to make you think about your data ... 

```python
#The following code plots the pixel mean and std for images in each class
plt.figure()
plt.plot(np.mean(images_red[labels_red==index_1,:],0),label='Cars')
plt.plot(np.mean(images_red[labels_red==index_2,:],0),'r',label='Trucks')
plt.legend()
plt.xlabel('Pixel')
plt.ylabel('Pixel Mean')
plt.title('Pixel Mean in Train Set')

plt.figure()
plt.plot(np.std(images_red[labels_red==index_1,:],0),label='Cars')
plt.plot(np.std(images_red[labels_red==index_2,:],0),'r',label='Trucks')
plt.legend()
plt.xlabel('Pixel')
plt.ylabel('Pixel Std')
plt.title('Pixel Std in Train Set')

```

In both graphs we can see that trucks have higher mean and standard deviation compared to cars, in particular in the first features of the pixels of each color. We can try to create and add a dummy variable which takes value 1 if the mean and standard deviation is greater than a certain value and 0 otherwise.<br>
Let's start by creating a new set of data, including the new variable:

```python
new = np.zeros([1955,3073])
newTest = np.zeros([2000,3073])

new[:,:-1] = images_red
newTest[:,:-1] = images_red_test

for index, i in enumerate(images_red[:,]):
    if np.mean(i) > 145 and np.std(i) > 75:
        new[i,-1] = 1 
        #print(new)
        
for index, i in enumerate(images_red_test[:,]):
    if np.mean(i) > 145 and np.std(i) > 75:
        newTest[i,-1] = 1 
        #print(newTest)
```

Then our new data have to be normalized before applying the SVC.

```python
new_norm, mu, std = normalize(new.astype(np.float64),mu=0,std=1,flag_train=True)
new_test_norm,_,_ = normalize(newTest.astype(np.float64),mu,std,flag_train=False)
```

Let's see if the two best models of the previous sections can be improved.

```python
best = SVC(C=10,kernel='rbf',gamma=0.5).fit(new_norm,labels_red)
p = best.predict(new_test_norm)
num = len(best.support_vectors_)

a =accuracy_score(labels_red_test, p)
e = 1-a
```

```python
print('The number of support vectors is:',num)
print('The misclassification error rate is:',e)
```

We can see that the SVC with the Gaussian kernel haven't improved.

```python
best = SVC(C=90,kernel='poly',degree=3, gamma=1).fit(new_norm,labels_red)
p = best.predict(new_test_norm)
num = len(best.support_vectors_)

a =accuracy_score(labels_red_test, p)
e = 1-a
print('The number of support vectors is:',num)
print('The misclassification error rate is:',e)
```

The polynomial kernel with degree 3 is worse in the third decimal and the SVC has 2 more support vectors compared to one classifier built without this new variable. <br>
Let's try if it could be worth to modify $\gamma$:

```python
def CVforGamma3(x,y,g):
    a=np.zeros([len(g),5])
    for index, i in enumerate(g):
        model = SVC(C=90,kernel='poly',gamma=i,degree=3)
        k=ShuffleSplit(5, 0.3)
        for train_index, test_index in k.split(x):
            fit = model.fit(x[train_index], y[train_index])
            scores = model.predict(x[test_index])
            a[index,] = accuracy_score(y[test_index],scores)
    return a  
```
