```python
def Read_Files(txt_files):
    
    files_list = []
    for file in txt_files:
        with open(file, 'r') as f:

            file = f.readlines()
            #print(file)
            files_list.append(file)
            
    return files_list

files_list = Read_Files(files)
```

```python
def Build_DataFrame(txt_files, files_list):
    df = pd.DataFrame()

    for index,file in enumerate(txt_files):

        filename = file.split('/')[-1]
        df.loc[index,'File'] = filename
        
    df['content'] = [''.join(x) for x in files_list]
    
    return df

df1 = Build_DataFrame(files, files_list); df1
```

```python
## apply string similarity to replace typos for certain words (ranges and ratio have to be specified, cannot be generalized)

def String_Similarities(word, df, ratio, interval):
    import Levenshtein as lev
    
    lev_found = []

    for index, content in enumerate(df['content']):
        split = content.split('.\n')
        for (idx_split,element) in enumerate(split):

            element = element.replace('\n', ' \n ')
            doc=nlp(element)
            matcher = Matcher(nlp.vocab)

            for (index2, token) in enumerate(element.split(' ')):
                if len(word) - interval < len(token) < len(word) + interval:
                    if word.startswith(token[0])==True:
                        if lev.ratio(word.lower(),token.lower()) > ratio:

                            print(word, token)
                            print(word.startswith(token[0]))
                            lev_found.append(token)
                            
    DF = pd.DataFrame()
    DF['change'] = [x for x in lev_found]
    DF['key'] = word; DF
    myDict = dict(zip(DF.change, DF.key)); myDict.keys()
    del myDict[word.upper()]; myDict.keys()
    
    df['correct'] = df['content'].replace(to_replace = myDict,regex=True)

    
    myDict = dict()
    myDict['Hospitall'] = 'Hospital'
    myDict
    df['content'] = df['correct'].replace(to_replace = myDict,regex=True)
    df = urgenciasDF.drop(['correct'], axis = 1)  
    
    return df
```

```python
def Disambiguation_Hospitals(hospitals, df, interval, ratio):
    ##  Function to do disambiguation of hospitals  
    out = []
    import Levenshtein as lev
    for hospital in hospitals:

        for (index, found) in enumerate(df['Hospital Name']):

            if df.loc[index,'Hospital Name'] != 'NaN':

                if (len(hospital) - interval) < len(str(found)) < (len(hospital) + interval):
                    if lev.ratio(hospital.lower(),str(found).lower()) > ratio:

                        #print(hospital, found)
                        out.append(found.rstrip())
                        
    return out            
```

```python
def Extract_DESPACHO(df):
    
    keywords = [ 'abogado', 'abogados', 'despacho', 'juris', 'juridico', 'juridicos', 'asesores']
    df['Despacho'] = ''
    for index, row in enumerate(df['Email']):
        
        for s, split in enumerate(str(row).lower().split('@')):

            for word in keywords:

                if word in split and word != split:

                    #print(str(str(split).split('.')[0]).split(word)[0] +' '+ word+' ' + str(str(split).split('.')[0]).split(word)[1])
                    df.loc[index, 'Despacho'] = str(str(str(split).split('.')[0]).split(word)[0] +' '+ word+' ' + str(str(split).split('.')[0]).split(word)[1]).title()
                    
                if word == split:
                    df.loc[index, 'Despacho'] = row.lower().replace('@', ' ').split('.')[0].title()
                    
                
                    
                    
    return df
```

```python
def Read_Files(txt_files):
    
    files_list = []
    for file in txt_files:
        with open(file, 'r') as f:

            file = f.readlines()
            print(file)
            files_list.append(file)
            
    return files_list

files_list = Read_Files(files); files_list[0]
```

```python
def Build_DataFrame(txt_files, files_list):
    df = pd.DataFrame()

    for index,file in enumerate(txt_files):

        filename = file.split('/')[-1]
        df.loc[index,'File'] = filename
        
    df['content'] = [''.join(x) for x in files_list]
    df['Line'] =''
    
    return df

mixDF = Build_DataFrame(files, files_list); mixDF
```

```python
def Add_Doctors(df):
    for index, (docts, content) in enumerate(zip(df['Doctors'],df['content'])):
        split = content.split('.\n')
        for j,element in enumerate(split):

            element = element.replace('\n', ' \n ')
            doc=nlp(element)
            matcher = Matcher(nlp.vocab)

            # Add the pattern to the matcher and apply the matcher to the doc
            matcher.add('sites', None, 
                        [{'POS':'PROPN', 'OP':'+'},{'TEXT':{'NOT_IN':'Fisioterapeuta.'}}, {'IS_ALPHA':False, 'OP':'?'}, {'TEXT':'NOCOLG'}], 
                        [{'POS':'PROPN', 'OP':'+'},{'TEXT':{'NOT_IN':'Fisioterapeuta.'}}, {'IS_ALPHA':False, 'OP':'?'}, {'TEXT':'No'}, {'TEXT':{'REGEX':'^Coleg'}}],
                        [{'POS':'PROPN', 'OP':'+'},{'TEXT':{'NOT_IN':'Fisioterapeuta.'}}, {'IS_ALPHA':False, 'OP':'?'}, {'TEXT':'No'}, {'TEXT':{'REGEX':'^Coleg'}}])

            matches = matcher(doc)
            print('Total matches found:', len(matches))
            if len(matches) != 0:
                # Iterate over the matches and print the span text
                for match_id, start, end in matches:
                    print('Match found:', doc[start:end].text)

                    
                    ## add NO rule!!!!!!!!!!!!!!!!!!!!!!!!
                    if 'No' in doc[start:end].text or if 'NO' in doc[start:end].text:
                        #df.loc[index, 'Med'] = doc[start:end].text.split(' ')[:-2]
                        print(' '.join(doc[start:end].text.split(' ')[:-2]).rstrip())

                        if ' '.join(doc[start:end].text.split(' ')[:-2]).rstrip() not in docts:
                            df.loc[index, 'Med'] = ' '.join(doc[start:end].text.split(' ')[:-2]).rstrip()

                            df.loc[index, 'Doctors'] |= {' '.join(doc[start:end].text.split(' ')[:-2]).rstrip()}


                    else:
                        if ' '.join(doc[start:end].text.split(' ')[:-1]).rstrip() not in docts:
                            print(' '.join(doc[start:end].text.split(' ')[:-1]).rstrip())
                            df.loc[index, 'Med'] = ' '.join(doc[start:end].text.split(' ')[:-1]).rstrip()
                            df.loc[index, 'Doctors'] |= {mixDF.loc[index, 'Med']}
                            
    def substringSieve(string_list):
        out = []
        for s in string_list:
            if not any([s in r for r in string_list if s != r]):
                out.append(s.rstrip())
        return out
    
    for index, row in enumerate(df['Doctors']):
        #print(substringSieve(row))
        df.loc[index, 'Doctors'] = ' // '.join(substringSieve(row))
        
    df = df.drop('Med', axis = 1)

    return df
                        
                    
```
