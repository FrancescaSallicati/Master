__SOLUTION__<br>
I chose to work on the Spanish eletrical system data and to use a value for the parameters of the penalty term equal to 1.<br>

By derivating the function in terms of $\beta$ we obtain:

$$\frac{\delta}{\delta \beta}((Y - X\beta)'(Y-X\beta) + \rho \beta'\beta))=-2X'(Y-\beta' X) + 2\rho\beta = 0 $$ 
$$ X'Y = X'X\beta + \rho \beta$$ 
$$\beta = (X'X+ \rho I )^{-1}X' Y$$

whose minimum is given by: $$\beta = (X'X+ \rho I )^{-1}X' Y$$


```python
%matplotlib inline
import numpy as np
import time
import pandas as pd
import matplotlib.pyplot as plt

price=pd.read_csv('price.csv',delimiter=';',usecols=['value'])
coal=pd.read_csv('coal.csv',delimiter=';',usecols=['value'])
gas=pd.read_csv('gas.csv',delimiter=';',usecols=['value'])
wind=pd.read_csv('wind.csv',delimiter=';',usecols=['value'])
hidro=pd.read_csv('hidro.csv',delimiter=';',usecols=['value'])
nuclear=pd.read_csv('nuclear.csv',delimiter=';',usecols=['value'])
solar=pd.read_csv('solar.csv',delimiter=';',usecols=['value'])
cogen=pd.read_csv('cogen.csv',delimiter=';',usecols=['value'])
demand=pd.read_csv('demand.csv',delimiter=';',usecols=['value'])
df=pd.concat([price, coal, gas, wind, hidro, nuclear, solar, cogen, demand],axis=1)
df.columns=['price', 'coal', 'gas', 'wind', 'hidro', 'nuclear', 'solar', 'cogen', 'demand']

normalized_df=(df-df.mean())/df.std()
normalized_df
```

__SOLUTION__<br>
Now we're going to use different types of solvers (such as Nelder-Mead, Powell, dogleg and so on) in order to see which one best suits our problem.

```python
from scipy.optimize import minimize
help(minimize)
```

```python
#ObjectiveFunction to minimize
def ridge_reg(beta_ls, X, Y,rho):
    beta_ls=np.matrix(beta_ls)
    z=Y-np.dot(X,beta_ls.T)
    return np.dot(z.T,z) + rho*np.dot(beta_ls,beta_ls.T)
```

And we will also need to define functions to obtain the gradient and the hessian, whose expressions are:<br>

__Gradient:__
$$ -2X' Y+2(X' X+\rho I)\beta $$
__Hessian:__
$$2(X'X+\rho\,I)$$

```python
#Gradient:
def ridge_reg_der(beta_ls,X,Y,rho):
    beta_ls=np.matrix(beta_ls)
    pp=-2*np.dot((Y-np.dot(X,beta_ls.T)).T,X)+2*rho*beta_ls
    aa=np.squeeze(np.asarray(pp))
    return aa

#Hessian:
def ridge_reg_hess(beta_ls,X,Y,rho):
    return 2*np.dot(X.T,X)+2*rho*np.identity(b)
```

Let's start by __Nelder_Mead__ method: 

```python
beta_ls0 = np.zeros(b)
time_start1 = time.clock()

res1 = minimize(ridge_reg,beta_ls0,args=(X,Y,rho),method="Nelder-Mead",
                            options={"disp":True,"xtol":1e-10}) 
time_elapsed1 = (time.clock() - time_start1)

print('Time',time_elapsed1)
print (res1.x) 
print('Error',np.linalg.norm(beta_ls_exact.T-res1.x,ord=2)/np.linalg.norm(beta_ls_exact.T,ord=2))
```

__c) Estimate the value of the regression coefficients by implementing the:<br>
i. (1.5 points) Gradient method.<br>
ii. (1.5 points) Newton method.<br>
Consider a line search technique to improve the algorithm convergence, e.x., Armijo rule. <br>
Compare the performance of these algorithms (number of iterations and total computational
time).__

__SOLUTION__<br>
__i.) Gradient Method (with Wolfe Conditions)__ <br>

$\bullet$ Initial iteratation $x_0$

$\bullet$ Computes search (descent) directions $p_k=-\nabla f(x_k)$

$\bullet$ Far from the solution, computes a steplength $\alpha_k>0$, to which Armijo rule is applied:

$$ f(x_k+\alpha_k p_k)\leq f(x_k)+c_1 \alpha_k\nabla f(x_k)' p_k$$   

$\bullet$ Movement:
$$x_{k+1} = x_k + \alpha_k\ p_k$$
Until convergence to a local solution.

```python
import time
(a,b)=X.shape
beta_lsg=np.zeros(b)
n_iter=5000
OF_iter=np.zeros(n_iter)
tol_iter=np.zeros(n_iter)
alpha_iter=np.zeros(n_iter)
i=0;
tol=10000;
epsilon=1e-3;

time_start7 = time.clock()
while (i <= n_iter-2) and (tol>epsilon):
    i=i+1
    grad= ridge_reg_der(beta_lsg,X,Y,rho)
    ddirect=-grad
    # Armijo condition
    #######################################
    sigma=0.1
    beta=0.1
    alpha=1
    while (ridge_reg(beta_lsg+alpha*ddirect,X,Y,rho)>
           ridge_reg(beta_lsg,X,Y,rho)+alpha*sigma*np.dot(ddirect,grad)) :
        alpha=alpha*beta
    #######################################
    beta_lsg= beta_lsg +alpha*ddirect
    OF_iter[i]=ridge_reg(beta_lsg, X, Y,rho)
    tol=np.linalg.norm(grad,ord=2)
    tol_iter[i]=tol
    alpha_iter[i]=alpha

time_elapsed7 = (time.clock() - time_start7) 
print('Time:',time_elapsed7)
print('Iterations:',i)
print('ObjectiveFunction value',OF_iter[i])
print("Computed Betas:",beta_lsg)
print("Exact Betas:",np.transpose(beta_ls_exact))
print('Tolerance:',tol)
print('Error',np.linalg.norm(np.transpose(beta_ls_exact)-beta_lsg,ord=2)/np.linalg.norm(beta_lsg,ord=2))
```

__i.) Newton Method (with Wolfe Conditions)__ <br>

$\bullet$ Initial iteratation $x_0$

$\bullet$ Computes search (descent) directions  $p_k=-(\nabla^2 f(x_k))^{-1} \nabla f(x_k)$ where $\nabla^2 f(x_k)$ is
nonsingular. 

$\bullet$ Far from the solution, computes a steplength $\alpha_k>0$, to which Armijo rule is applied:

$$ f(x_k+\alpha_k p_k)\leq f(x_k)+c_1 \alpha_k\nabla f(x_k)' p_k$$   

$\bullet$ Movement:
$$x_{k+1} = x_k + \alpha_k\ p_k$$
Until convergence to a local solution.

```python
import time
(a,b)=X.shape
beta_lsn=np.zeros(b) 
n_iter=5000 
OF_iter2=np.zeros(n_iter)
tol_iter=np.zeros(n_iter)
alpha_iter=np.zeros(n_iter)
j=0;
tol=10000;
epsilon=1e-3;

time_start8 = time.clock()
while (j <= n_iter-2) and (tol>epsilon):
    j=j+1
    grad= ridge_reg_der(beta_lsg,X,Y,rho)
    hess= ridge_reg_hess(beta_lsn,X,Y,rho)
    ddirect= -np.dot(inv(hess),grad)
     # Armijo condition (opcionales - tamaño de paso variable)
    #######################################
    sigma=0.1
    beta=0.1
    alpha=1
    while (ridge_reg(beta_lsg+alpha*ddirect,X,Y,rho)>
           ridge_reg(beta_lsg,X,Y,rho)+alpha*sigma*np.dot(ddirect,grad)) :
        alpha=alpha*beta
    beta_lsg= beta_lsg +alpha*ddirect
    OF_iter2[j]=ridge_reg(beta_lsg, X, Y,rho)
    tol=np.absolute((OF_iter2[j]-OF_iter2[j-1])/OF_iter2[j]) #tolerance
    tol=np.linalg.norm(grad,ord=2)
    tol_iter[j]=tol
    alpha_iter[j]=alpha

time_elapsed8 = (time.clock() - time_start8) 
print('Time',time_elapsed8)
print('Iterations',j)
print('ObjectiveFunction value',OF_iter2[j])
print("Computed Betas:",beta_lsg)
print("Exact Betas",np.transpose(beta_ls_exact))
print('Tolerance=',tol)
print('Error',np.linalg.norm(np.transpose(beta_ls_exact)-beta_lsg,ord=2)/np.linalg.norm(beta_lsg,ord=2))
```

```python
time=[time_elapsed7,time_elapsed8]
fun=[OF_iter[i],OF_iter2[j]]
it=[i,j]
print('Objective Function Value:',fun)
print('Iterations:',it)
print('Time:',time)
```

We can see that Netwton's methods performs faster than the Gradient descending method due to the number of iterations which is 2 compared to the 199 of the Gradient method.

__d) Estimate the value of the regression coefficients by implementing the:<br>
i. Coordinate gradient method.<br>
ii. Stochastic gradient method.__<br>

__SOLUTION:<br>
i. Coordinate gradient method.__<br>

$\bullet$  From an initial iterate $x_0$

$\bullet$ Compute search (descent) directions. Find $i_k \in {1,...,p}$ such that $\nabla_{i_k} f(x_k)$ is maximum. $p_k=- \nabla_{i_k} f(x_k) e_{i_k}$

$\bullet$ Far from the solution, compute a steplength $\alpha_k>0$:

$\bullet$ Movement:
$$x_{k+1} = x_k + \alpha_k\ p_k$$
Until convergence to a local solution.


```python
import time
(a,b)=X.shape
beta_lsg=np.zeros(b) 

alpha=0.0001
n_iter=5000 
OF_iter3=np.zeros(n_iter)
tol_iter=np.zeros(n_iter)
alpha_iter=np.zeros(n_iter)
k=0;
tol=10000;
epsilon=1e-3;

time_start9 = time.clock()
while (k <= n_iter-2) and (tol>epsilon):
    k=k+1
    grad= ridge_reg_der(beta_lsg,X,Y,rho)
    ddirect=-np.eye(1,len(grad),np.argmax(abs(grad)))*grad
    
    beta_lsg= beta_lsg +alpha*ddirect
    OF_iter3[k]=ridge_reg(beta_lsg, X, Y,rho)
    tol=np.linalg.norm(grad,ord=2)
    tol_iter[k]=tol
    alpha_iter[k]=alpha
    

time_elapsed9 = (time.clock() - time_start9) 
print('Time',time_elapsed9)
print('Iterations',k)
print('ObjectiveFunction value',OF_iter3[k])
print("Computed Betas:",beta_lsg)
print("Exact Betas",np.transpose(beta_ls_exact))
print('Tolerance=',tol)
print('Error',np.linalg.norm(np.transpose(beta_ls_exact)-beta_lsg,ord=2)/np.linalg.norm(beta_lsg,ord=2))
```

__ii. Stochastic gradient method.__<br>

$\bullet$ From an initial iterate $x_0$

$\bullet$ Compute search (descent) directions. Choose $j_k \in {1,...,n}$ at random $p_k=- \nabla f_{j_k}(x_k) $

$\bullet$ Far from the solution, compute a steplength $\alpha_k>0$:

$\bullet$ Movement:
$$x_{k+1} = x_k + \alpha_k\ p_k$$
Until convergence to a local solution.

```python
import random

#Random Gradient:
def ridge_reg_der2(beta_ls,X,Y,rho):
    j=int(random.uniform(1,len(Y))) 
    beta_ls=np.matrix(beta_ls)
    pp=-2*(Y[j]-np.dot(X[j,:],beta_ls.T)).T*X[j,:]+2*rho*beta_ls
    aa=np.squeeze(np.asarray(pp))
    return aa
```

```python
(a,b)=X.shape
beta_lsg=np.zeros(b) 

alpha=0.0001
n_iter=5000 
OF_iter4=np.zeros(n_iter)
tol_iter=np.zeros(n_iter)
alpha_iter=np.zeros(n_iter)
z=0;
tol=10000;
epsilon=1e-3;

time_start10 = time.clock()
while (z <= n_iter-2) and (tol>epsilon):
    z=z+1
    grad= ridge_reg_der2(beta_lsg,X,Y,rho)
    ddirect=-grad
    beta_lsg= beta_lsg +alpha*ddirect
    OF_iter4[z]=ridge_reg(beta_lsg, X, Y,rho)
    tol=np.linalg.norm(grad,ord=2)
    tol_iter[z]=tol
    alpha_iter[z]=alpha
    

time_elapsed10 = (time.clock() - time_start10) 
print('Time',time_elapsed10)
print('Iterations',z)
print('ObjectiveFunction value',OF_iter4[z])
print("Computed Betas:",beta_lsg)
print("Exact Betas",np.transpose(beta_ls_exact))
print('Tolerance=',tol)
print('Error',np.linalg.norm(np.transpose(beta_ls_exact)-beta_lsg,ord=2)/np.linalg.norm(beta_lsg,ord=2))
```

```python
time=[time_elapsed9,time_elapsed10]
fun=[OF_iter3[k],OF_iter4[z]]
it=[k,z]
print('Objective Function Value:',fun)
print('Iterations:',it)
print('Time:',time)
```

The results given by Coordinate Gradient method are better in terms of fewer iterations (2050 vs 4999 of the Stochastic method) and time elapsed. Moreover the error given by the Stochastic method is much higher and the algorithm actually converges to a different minimum compared to all the previous methods.
